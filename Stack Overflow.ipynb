```python
selected_column=['Country','Age','Gender','EdLevel','UndergradMajor',
                'Hobbyist','Age1stCode','YearsCode','YearsCodePro','LanguageWorkedWith',
                'LanguageDesireNextYear','NEWLearn','NEWStuck',
                'Employment','DevType',
                'WorkWeekHrs','JobSat','JobFactors','NEWOvertime','NEWEdImpt']
```

```python
len(selected_column)
```

```python
survey_df=survey_raw_df[selected_column].copy()
```

```python
survey_df
```

```python
schema=survey_raw[selected_column]
```

```python
schema.shape
```

```python
survey_df.shape
```

```python
survey_df.info()
```

```python
survey_df['Age1stCode']=pd.to_numeric(survey_df.Age1stCode, errors='coerce')
survey_df['YearsCode']=pd.to_numeric(survey_df.YearsCode, errors='coerce')
survey_df['YearsCodePro']=pd.to_numeric(survey_df.YearsCodePro, errors='coerce')
```

```python
survey_df.describe()
```

```python
survey_df.drop(survey_df[survey_df.Age<10].index,inplace=True)
survey_df.drop(survey_df[survey_df.Age>100].index,inplace=True)
```

```python
survey_df.drop(survey_df[survey_df.WorkWeekHrs>140].index,inplace=True)
```

```python
survey_df.Gender.value_counts()
```

```python
survey_df.where(~(survey_df.Gender.str.contains(';',na=False)),np.nan,inplace=True)
```

```python
survey_df.Gender.value_counts()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['figure.facecolor']='#00000000'
```

```python
survey_df.Country.nunique()
```

```python
top_countries=survey_df.Country.value_counts().head(15)
top_countries
```

```python
plt.figure(figsize=(12,9))
plt.xticks(rotation=90)
plt.title(schema.Country)
sns.barplot(x=top_countries.index,y=top_countries);
```

```python
plt.figure(figsize=(12,9))
plt.title(schema.Age)
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.hist(survey_df.Age,bins=np.arange(10,100,5), color="purple");
```

```python
schema.Gender
```

```python
gender_counts=survey_df.Gender.value_counts(dropna=False)
gender_counts
```

```python
plt.figure(figsize=(12,9))
plt.title(schema.Gender)
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%');
```

```python
plt.figure(figsize=(12,12))
plt.title(schema.EdLevel)
sns.countplot(y=survey_df.EdLevel)
plt.xlabel("Count")
plt.ylabel(None);
```

```python
schema.UndergradMajor
```

```python
UndergradMajor_pct=survey_df.UndergradMajor.value_counts()/survey_df.UndergradMajor.count()*100
```

```python
plt.figure(figsize=(12,9))
plt.ylabel(None);
plt.title(schema.UndergradMajor)
sns.barplot(x=UndergradMajor_pct,y=UndergradMajor_pct.index)
plt.xlabel("Percentage");
```
